#### **Handling Missing Value**

Ada istilah dalam dunia data "Garbage in Garbage Out", yang artinya jika data yang kita analisa masih memiliki sampah, maka hasil analisa kita juga 'sampah' atau tidak berkualitas. Salah satu bagian dalam data cleaning adalah menangani data yang kosong (missing values), yang dapat dilakukan dengan 2 cara: 
1. **Mengisi data yang hilang**: pengisian data yang hilang dapat dilakukan dengan berbagai cara, yang paling umum adalah dengan menggunakan data dari kolom lain yang secara domain knowledge atau statistik berkaitan dengan kolom yang kosong. Dalam analisa ini, beberapa kolom yang kosong memiliki keterkaitan dengan kolom lain, sehingga pengisian missing value dapat digunakan dengan metode ini.
2. **Menghapus data**: Jika setelah dilakukan pengisian dengan kolom lain masih menyisakan missing value, maka cara terakhir yang dapat digunakan adalah dengan menghapus data 

##### **CorridorID**: Mengisi data yang kosong pada kolom corridorID

```python
# Mengisi kolom corridorID dengan acuan dari kolom corridorName
df["corridorID"] = df["corridorID"].fillna(df.groupby("corridorName")["corridorID"].transform("first"))

# Mengisi kolom corridorID dengan acuan dari kolom tapInStopsName
df["corridorID"] = df["corridorID"].fillna(df.groupby("tapInStopsName")["corridorID"].transform("first"))

# Mengisi kolom corridorID dengan acuan dari kolom tapOutStopsName
df["corridorID"] = df["corridorID"].fillna(df.groupby("tapOutStopsName")["corridorID"].transform("first"))

# Mengisi kolom corridorID dengan acuan dari kolom payCardName
df["corridorID"] = df["corridorID"].fillna(df.groupby("payCardName")["corridorID"].transform("first"))

df.dropna(subset=['corridorID'], inplace=True)
```

Dalam dataset ini, missing values pada kolom `corridorID` dapat diisi dengan beberapa kolom yang berkaitan secara domain knowledge. Dan kolom yang paling berkaitan adalah kolom `corridorName`, sehingga pengisian missing values dilakukan dengan menggunakan referensi data dari kolom corridorName.
- Namun, setelah dilakukan pengisian missing values dari kolom `corridorName`, masih teradapat 1.125 data `corridorID` yang kosong, sehingga pengisian data kosong perlu dilanjutkan dengan acuan data dari kolom lain yang masih berkaitan, yakni kolom `TapInStopsName`
- Setelah diisi dengan acuan data dari kolom `TapInStopsName`, masih teradapat 33 data `corridorID` yang kosong, sehingga pengisian data kosong dilanjutkan dengan acuan data dari kolom `TapOutStopsName` 
- Setelah diisi dengan acuan data dari kolom `TapOutStopsName`, masih teradapat 17 data `corridorID` yang kosong, sehingga pengisian data kosong dilanjutkan dengan acuan data dari `PayCardName`  
- Setelah diisi dengan acuan data dari kolom `PayCardName`, masih teradapat 1 data corridorID yang kosong, sehingga **data kosong dihapus**    

##### **CorridorName**: Mengisi data yang kosong pada kolom corridorName

Pengisian missing values pada kolom `corridorID` sebelumnya tidak langsung terisi semua dengan data dari kolom `corridorName` karena pada kolom ini juga terdapat data yang masih kosong. Oleh karena itu pengisian missing value pada kolom ini dilakukan dengan data dari `corridorID` yang telah lengkap

```python
# Mengisi kolom corridorName dengan acuan dari kolom corridorID
df["corridorName"] = df["corridorName"].fillna(df.groupby("corridorID")["corridorName"].transform("first"))
```

```python
# Mengisi kolom tapInStops dengan acuan dari kolom tapInStopsName
df['tapInStops'] = df.groupby('tapInStopsName')['tapInStops'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x)
```

Ketentuan ini menjadi dasar untuk mengisi kolom kosong pada tabel payAmount ini:

**1. Mengisi missing value payAmount dengan 0 untuk lansia di atas 60 tahun, berdasarkan kolom payCardbirthDate <1962**

```python
# Mengecek jumlah missing value berdasarkan usia yang di atas 60 tahun dari kolom payCardBirthDate
df[df["payCardBirthDate"]<=1962].groupby("payCardBirthDate")["payAmount"].apply(lambda x: x.isna().sum())
```

```python
# Mengisi missing value payAmount dengan 0 untuk data dengan payCardBirtDate <=1962 atau usia 60 tahun ke atas
df.loc[df["payCardBirthDate"]<=1962, "payAmount"] = df.loc[df["payCardBirthDate"]<=1962, "payAmount"].fillna(0)
```

**2. Mengisi missing value berdasarkan corridorID**

Transjakarta memiliki 4 tipe layanan dimana masing-masing layanan memiliki ID unik untuk nomor koridor / rutenya, yang dalam dataset ini ada di kolom corridorID. Sehingga pengisian missing value yang pertama dapat dilakukan adalah dengan payAmount / Tarif masing-masing layanan.

```python
# Mengecek jumlah missing value payAmount berdasarkan kolom corridorID
df.groupby("corridorID")["payAmount"].apply(lambda x: x.isna().sum()).loc[lambda x: x > 0]
```

Setelah dilakukan pengecekan missing value pada kolom payAmount berdasarkan corridorID-nya, ditemukan:
- 2 corridorID dengan kode JAK yang merupakan mikrotrans yakni JAK.48A dan JAK.48B. Tarif Mikrotrans adalah Rp 0 / gratis
- 1 corridorID bus premium yakni B13 yang memiliki tarif Rp 20000
- 9 corridorID sisanya merupakan bus reguler dengan tarif 3500.

Maka pengisian missing value pada kolom payAmount dilakukan dengan ketentuan corridorID-nya sebagaimana yang telah disebutkan.

```python
# Mengisi missing value payAmount berdasarkan kolom corridorID JAK.48A dan JAK.48B dengan Rp 0
df["payAmount"] = df["payAmount"].fillna(df["corridorID"].apply(lambda x: 0 if x in ["JAK.48A", "JAK.48B"] else np.nan))

# corridorID bis premium diisi dengan Rp 20000, dan sisanya yang merupakan layanan bus reguler diisi dengan value 3500
df["payAmount"] = df["payAmount"].fillna(df["corridorID"].apply(lambda x: 20000 if x in ["1K","1T","6P","B13","B14","D31","D32","S12","S13","S14","S31","T21"] else 3500))
```

Selain merubah tipe data, dilakukan juga perubahan isi data agar dapat lebih mudah dipahami serta mempermudah analisa dan interpretasi, yakni dengan merubah:
- Kolom `payCardSex` yang awalnya berisi M dan F, diubah menjadi "Pria" untuk M, dan "Wanita" untuk F 
- Data pada kolom `direction` diubah dari 0 menjadi 'pergi' dan 1 menjadi 'pulang'
- Data kolom `payCardBank` diubah dengan menambah nama bank dan merapikan tulisannya

```python
# Merubah data jenis kelamin menjadi "Pria" dan "Wanita"
df["payCardSex"] = df["payCardSex"].apply(lambda x: "Pria" if x == "M" else "Wanita")

# Merubah data direction menjadi Pergi dan Pulang
df['direction'] = df['direction'].replace({0: 'Pergi', 1:'Pulang'})

# Mengubah penamaan data payCardBank dengan menambahkan nama bank dan merapikan tulisannya
df['payCardBank'] = df['payCardBank'].replace({'emoney': 'Mandiri e-Money', 'dki': 'JakCard', 'flazz': 'BCA Flazz', 'online': 'Online payment', 'brizzi': 'BRI Brizzi', 'bni': 'BNI TapCash'})
```

#### **Data Creation**

Setelah mengubah type data dan penamaan data, selanjutnya dibutuhkan beberapa data untuk analisa yang bisa dibuat dengan memanfaatkan informasi dari kolom-kolom data yang ada, sehingga diputuskan untuk membuat kolom-kolom baru berikut:
1. Kolom perjalanan per jam: `tripHour`, 
2. Kolom perjalanan per hari: `tripDay`, 
3. Kolom perjalanan per minggu: `tripWeeks`
4. Kolom perjalanan per bulan: `tripMonth`
5. Kolom durasi perjalanan dalam format (jam:menit:detik): `trip_duration` 
6. Kolom durasi perjalanan dalam menit: `trip_duration_minute`
7. Kolom usia: `age`
8. Kolom kategori usia: `ageCategory`
9. Kolom tipe penumpang: `passengerType`
10. Kolom tipe layanan transjakarta: `tipeLayanan`

```python
# 1. Kolom perjalanan per jam
df["tripHour"] = df["tapInTime"].dt.hour

# 2. Kolom perjalanan per hari
df["tripDay"] = df["tapInTime"].dt.day_name()

# 3. Kolom perjalanan per minggu
df["tripWeeks"] = df["tripDay"].apply(lambda x: "Weekend" if x in ["Saturday", "Sunday"] else "Weekday")

# 5. Kolom durasi perjalanan dalam format (jam:menit:detik)
df["trip_duration"] = df["tapOutTime"] - df["tapInTime"]

# 6. Kolom durasi perjalanan dalam menit
df['trip_duration_minute'] = (df['trip_duration'].dt.total_seconds()/60).round(2)

# 7. Kolom usia
df["age"] = df["tapInTime"].dt.year - df["payCardBirthDate"]
```

8. Kolom kategori usia

```python
# Mengecek usia minimum dan maksimum
min_usia = df["age"].min()
max_usia = df["age"].max()
print(f'Usia minimum dalam dataset: {min_usia} tahun')
print(f'Usia maksimal dalam dataset: {max_usia} tahun')
```

```python
# Fungsi untuk kategorisasi usia
def kategori_usia(df, kolom_usia):
    def kategori(usia):
        if 11 <= usia <= 25:
            return "Gen Z"
        elif 26 <= usia <= 40:
            return "Millenial"
        elif 41 <= usia <= 59:
            return "Gen X"
        elif 60 <= usia <= 77:
            return "Baby Boomers"
        else:
            return "Usia tidak valid"
    
    df["ageCategory"] = df[kolom_usia].apply(kategori)
    return df

# Menambahkan kolom ageCategory
df = kategori_usia(df, "age")
```

9. Kolom tipe penumpang

```python
df['tripMonth'] = df['tapInTime'].dt.to_period('M')
df['tripDate'] = df['tapInTime'].dt.date
frekuensiPerjalanan = df.groupby(['payCardID', 'tripMonth'])['tripDate'].nunique().reset_index()

def kategori_frekuensi(days):
    if days > 15:
        return 'Penumpang Aktif'
    elif days > 5:
        return 'Penumpang Reguler'
    else:
        return 'Penumpang Sesekali'

frekuensiPerjalanan['passengerType'] = frekuensiPerjalanan['tripDate'].apply(kategori_frekuensi)

df = df.merge(frekuensiPerjalanan[['payCardID', 'tripMonth', 'passengerType']], on=['payCardID', 'tripMonth'], how='left')

df.drop(columns=['tripDate', 'tripMonth'], inplace=True)
```

```python
# Analisa IQR
subset = df['age']
# Menghitung Q1, Q3, dan IQR
Q1 = np.percentile(subset, 25)
Q2 = np.percentile(subset, 50)
Q3 = np.percentile(subset, 75)
IQR = Q3 - Q1
    
# Menghitung batas bawah dan batas atas
lower_bound = round(Q1 - (1.5 * IQR), 2)
upper_bound = round(Q3 + (1.5 * IQR), 2)

outliers_data = []
jumlah_outliers = ((subset < lower_bound) | (subset > upper_bound)).sum()
outliers_data.append({
        'Q1': round(Q1, 2),
        'Q2': round(Q2, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Batas Bawah': lower_bound,
        'Batas Atas' : upper_bound,
        'Jumlah Outliers': jumlah_outliers})
df_ageOutliers = pd.DataFrame(outliers_data)
df_ageOutliers
```

**Identifikasi Outliers & Analisa IQR**

```python
sns.boxplot(data=df, x='tipeLayanan', y='age', hue='tipeLayanan', palette=warna_tj)

plt.title("Boxplot Usia Penumpang per Tipe Layanan", fontweight='bold', pad=15)
plt.xlabel('Tipe Layanan', fontweight='bold')
plt.ylabel('Usia Penumpang', fontweight='bold')
```

```python
outliers_data = []

for layanan in df['tipeLayanan'].unique():
    subset = df[df['tipeLayanan'] == layanan]['age']
    
    # Menghitung Q1, Q3, dan IQR
    Q1 = np.percentile(subset, 25)
    Q2 = np.percentile(subset, 50)
    Q3 = np.percentile(subset, 75)
    IQR = Q3 - Q1
    
    # Menghitung batas bawah dan batas atas
    lower_bound = round(Q1 - (1.5 * IQR), 2)
    upper_bound = round(Q3 + (1.5 * IQR), 2)
    
    jumlah_outliers = ((subset < lower_bound) | (subset > upper_bound)).sum()

    outliers_data.append({
        'Tipe Layanan': layanan,
        'Q1': round(Q1, 2),
        'Q2': round(Q2, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Batas Bawah': lower_bound,
        'Batas Atas' : upper_bound,
        'Jumlah Outliers': jumlah_outliers})

df_ageByService_IQR = pd.DataFrame(outliers_data)
df_ageByService_IQR
```

```python
# Analisa IQR
subset = df['trip_duration_minute']
# Menghitung Q1, Q3, dan IQR
Q1 = np.percentile(subset, 25)
Q2 = np.percentile(subset, 50)
Q3 = np.percentile(subset, 75)
IQR = Q3 - Q1
    
# Menghitung batas bawah dan batas atas
lower_bound = round(Q1 - (1.5 * IQR))
upper_bound = round(Q3 + (1.5 * IQR))

outliers_data = []
jumlah_outliers = ((subset < lower_bound) | (subset > upper_bound)).sum()
outliers_data.append({
        'Q1': round(Q1),
        'Q2': round(Q2),
        'Q3': round(Q3),
        'IQR': round(IQR),
        'Batas Bawah': lower_bound,
        'Batas Atas' : upper_bound,
        'Jumlah Outliers': jumlah_outliers})
df_durationOutliers = pd.DataFrame(outliers_data)
display(Markdown('**Durasi Perjalanan (dalam Menit)**'))
display(df_durationOutliers)
```

**Insight:**
- Berdasarkan uji normalitas distribusi data, diketahui pvalue < 0.05 sehingga dapat dikatakan bahwa **data durasi perjalanan tidak terdistribusi normal**. 
- Karena distribusinya tidak normal, maka untuk pemusatan datanya digunakan Median (Q2), sehingga diketahui bahwa **rata-rata durasi perjalanan keseluruhan adalah 72 menit atau 1 jam 12 menit**.
- Rentang interkuartil (IQR) sebesar 45 menit menunjukkan variasi durasi dimana **50% perjalanan berdurasi antara 51-96 menit / 51 menit - 1 jam 36 menit.**
- Batas atas outlier adalah 163 menit, di mana terdapat 36 data **perjalanan dengan durasi lebih dari 163 menit yang dianggap outlier**.

**Kesimpulan:**

Sebagian besar perjalanan berdurasi antara 51 hingga 96 menit (IQR), dengan median durasi sebesar 72 menit (1 jam 12 menit). Hal ini menunjukkan bahwa mayoritas penumpang menempuh perjalanan dalam rentang waktu yang cukup stabil. Namun, terdapat 36 perjalanan dengan durasi di atas 163 menit yang dikategorikan sebagai outlier. Durasi ekstrem ini kemungkinan besar dipengaruhi oleh kondisi lalu lintas yang sangat padat atau gangguan operasional lainnya.

**Mann-Whitney Test Median Durasi Perjalanan**

```python
from scipy.stats import mannwhitneyu

durasi_weekday = df[df['tripWeeks'] == 'Weekday']['trip_duration_minute']
durasi_weekend = df[df['tripWeeks'] == 'Weekend']['trip_duration_minute']

stat, pvalue = mannwhitneyu(durasi_weekday, durasi_weekend)

if pvalue >= 0.05: #Ho
    print(f'''P-Value = {pvalue} > 0.05. Gagal Menolak Ho
          Kita TIDAK PUNYA cukup bukti untuk mengatakan bahwa median durasi perjalanan saat Weekday tidak sama dengan median durasi saat Weekend.
          (Rata-rata durasi perjalanan saat Weekday dan Weekend TIDAK BERBEDA signifikan)''')
else: #Ha
    print(f'''P-Value = {pvalue} < 0.05. Berhasil Menolak Ho
          Kita PUNYA cukup bukti untuk mengatakan bahwa median durasi perjalanan saat Weekday tidak sama dengan median durasi saat Weekend.
          (Rata-rata durasi perjalanan saat Weekday dan Weekend BERBEDA signifikan)''')
```

**Insight:**
- Median durasi perjalanan saat **weekday 72 menit**, saat **weekend 70 menit**.
- Secara praktis tidak ada perbedaan signifikan dalam durasi perjalanan rata-rata antara hari kerja dan hari libur. Namun secara statistik berdasarkan Mann Whitney Test, perbedaan 2 menit dikatakan sebagai perbedaan yang signifikan.
- Durasi rata-rata saat weekday yang sedikit lebih lama, kemungkinan disebabkan karena arus lalu lintas yang lebih padat.
- IQR **weekend 50 menit** lebih besar dari IQR **weekday 44 menit**
- Nilai IQR menunjukkan bahwa variasi durasi perjalanan saat weekend lebih tinggi, artinya **waktu tempuh perjalanan saat akhir pekan lebih tidak stabil atau lebih beragam**, kemungkinan karena rute perjalanan yang lebih bervariasi (misalnya ke tempat wisata), perubahan perilaku pengguna, atau variasi kepadatan lalu lintas yang tidak menentu.
- Jumlah **outlier saat weekday (28) jauh lebih banyak** dibandingkan weekend (8).
- Ini menunjukkan bahwa p**erjalanan ekstrem berdurasi lama lebih sering terjadi di hari kerja**, yang bisa jadi akibat dari kemacetan parah, keterlambatan operasional, atau rute panjang yang digunakan oleh penumpang saat weekday.

```python
# Fungsi untuk membuat dataframe analisa IQR tiap Layanan
def hitung_iqr_per_week(df_input, tipe_layanan):
    outliers_data = []

    for weeks in df_input['tripWeeks'].unique():
        subset = df_input[df_input['tripWeeks'] == weeks]['trip_duration_minute']

        # Hitung Q1, Q2, Q3, dan IQR
        Q1 = np.percentile(subset, 25)
        Q2 = np.percentile(subset, 50)
        Q3 = np.percentile(subset, 75)
        IQR = Q3 - Q1

        # Hitung batas bawah dan batas atas
        lower_bound = round(Q1 - 1.5 * IQR)
        upper_bound = round(Q3 + 1.5 * IQR)

        jumlah_outliers = ((subset < lower_bound) | (subset > upper_bound)).sum()

        outliers_data.append({
            'Trip Week': weeks,
            'Q1': round(Q1),
            'Q2': round(Q2),
            'Q3': round(Q3),
            'IQR': round(IQR),
            'Batas Bawah': lower_bound,
            'Batas Atas': upper_bound,
            'Jumlah Outliers': jumlah_outliers
        })

    # Kembalikan hasil sebagai dataframe
    return pd.DataFrame(outliers_data)
```

```python
durationPerWeeks_BRT = hitung_iqr_per_week(df_brt, 'BRT')
durationPerWeeks_NonBRT = hitung_iqr_per_week(df_nonBRT, 'NonBRT')
durationPerWeeks_RoyalTrans = hitung_iqr_per_week(df_royalTrans, 'RoyalTrans')
durationPerWeeks_Mikrotrans = hitung_iqr_per_week(df_mikrotrans, 'Mikrotrans')

display(Markdown('**IQR Durasi Perjalanan BRT (dalam menit)**'))
display(durationPerWeeks_BRT)
display(Markdown('**IQR Durasi Perjalanan Non-BRT (dalam menit)**'))
display(durationPerWeeks_NonBRT)
display(Markdown('**IQR Durasi Perjalanan RoyalTrans (dalam menit)**'))
display(durationPerWeeks_RoyalTrans)
display(Markdown('**IQR Durasi Perjalanan Mikrotrans (dalam menit)**'))
display(durationPerWeeks_Mikrotrans)
```

**Insight:**
- Median durasi perjalanan (Q2) sedikit lebih rendah saat weekend (1–3 menit lebih cepat), kemungkinan disebabkan oleh lalu lintas yang lebih lengang di akhir pekan karena aktivitas rutin seperti jam masuk kantor, sekolah, dan kepadatan jam pulang kerja berkurang.
- Variasi durasi perjalanan (IQR) sedikit lebih besar saat weekend, yang bisa disebabkan oleh pola perjalanan yang lebih fleksibel dan beragam. Saat weekend, penumpang mungkin melakukan perjalanan dengan tujuan hiburan, wisata, atau kegiatan pribadi yang menyebar ke berbagai waktu dan lokasi, tidak sepadat dan se-seragam weekday.
- Batas atas durasi perjalanan lebih tinggi saat weekend, mengindikasikan bahwa meskipun sebagian besar perjalanan berjalan lebih cepat, masih ada kemungkinan rute tertentu mengalami hambatan temporer, seperti kepadatan di area wisata, CFD, atau acara publik.

Dengan program ini, diharapkan penumpang aktif terdorong untuk tetap menggunakan layanan Transjakarta di akhir pekan, baik untuk keperluan rekreasi maupun aktivitas lainnya. Selain itu, insentif ini juga berpotensi menarik minat penumpang non-aktif (seperti penumpang reguler atau sesekali) untuk meningkatkan frekuensi penggunaan mereka demi mencapai status penumpang aktif dan memperoleh manfaat serupa. Melalui pendekatan ini, Transjakarta dapat memperluas keterlibatan pengguna setia sekaligus mendorong peningkatan jumlah perjalanan secara signifikan di akhir pekan.

Aktivitas perjalanan penumpang TransJakarta didominasi pada hari kerja (weekday), terutama di jam-jam sibuk yang puncaknya terjadi pada pukul 06:00 pagi dan 17:00 sore. Pola ini menunjukkan bahwa seluruh layanan TransJakarta merupakan moda transportasi andalan bagi para pekerja di Jakarta untuk mobilitas sehari-hari, khususnya pada hari kerja.

### **Kesimpulan: Analisa Pola Perjalanan Berdasarkan Rute**

Berdasarkan analisa top 10 corridor saat weekday vs weekend, dapat diketahui bahwa adanya perbedaan pola penggunaan transjakarta sebagai moda transportasi masyarakat saat hari kerja dan hari libur yang sangat dipengaruhi oleh kebutuhan dan aktivitas harian masyarakat. Perbedaan pola ini menunjukkan fleksibilitas Transjakarta dalam melayani berbagai kebutuhan mobilitas warga Jakarta dan sekitarnya, sekaligus menegaskan pentingnya penyesuaian operasional dan alokasi armada yang responsive terhadap pola permintaan yang berbeda antara hari kerja dan akhir pekan.

**2. Berdasarkan Layanan:**

Pola perjalanan berdasarkan rute tiap tipe layanan dapat dilihat bahwa layanan RoyalTrans dan Non-BRT banyak digunakan untuk perjalanan jarak menengah dan jauh dari wilayah penyangga seperti Cibubur, Bekasi, atau Ciputat ke pusat kota, terutama saat weekday. Perbedaan pola penggunaan tiap layanan menunjukkan bahwa setiap tipe layanan Transjakarta memiliki peran strategis dan saling melengkapi dalam menjawab kebutuhan mobilitas masyarakat yang beragam sesuai dengan waktu dan tujuan perjalanan.

### **Kesimpulan: Analisa Durasi Perjalanan**

**1. Secara Keseluruhan:**

Sebagian besar perjalanan berdurasi antara 51 hingga 96 menit (IQR), dengan median durasi sebesar 72 menit (1 jam 12 menit). Hal ini menunjukkan bahwa mayoritas penumpang menempuh perjalanan dalam rentang waktu yang cukup stabil. Namun, terdapat 36 perjalanan dengan durasi di atas 163 menit yang dikategorikan sebagai outlier. Durasi ekstrem ini kemungkinan besar dipengaruhi oleh kondisi lalu lintas yang sangat padat atau gangguan operasional lainnya.

**2. Berdasarkan Layanan:**
- **Layanan BRT** menunjukkan performa waktu perjalanan yang relatif stabil dengan median durasi sekitar 71 menit, serta rentang interkuartil (IQR) sebesar 43 menit. Artinya, separuh dari perjalanan berlangsung antara 51 hingga 94 menit, dengan ketidakpastian waktu tempuh sekitar ±20–25 menit dari rata-rata. Walaupun tergolong stabil, terdapat 7 kasus outlier dengan durasi ekstrem yang jauh melebihi batas wajar (di atas 158 menit), menunjukkan adanya potensi keterlambatan signifikan pada kondisi-kondisi tertentu seperti kepadatan lalu lintas atau gangguan operasional. Namun, secara keseluruhan, BRT tetap menjadi layanan yang cukup andal karena memiliki jalur khusus yang meminimalkan paparan terhadap kemacetan.

- **Layanan Non-BRT** memiliki median durasi yang hampir sama dengan BRT yaitu sekitar 72 menit, namun dengan tingkat ketidakpastian yang sedikit lebih tinggi (IQR 45 menit). Hal yang menonjol adalah tingginya jumlah outliers (21 kasus), menandakan bahwa variasi waktu perjalanan pada layanan ini cukup besar disebabkan karena 2 faktor, yaitu cakupan rute Non-BRT yang jauh lebih luas dibandingkan layanan lain, dan ketiadaan jalur khusus yang membuat Non-BRT lebih rentan terhadap kemacetan lalu lintas, sehingga waktu tempuh dapat meningkat drastis terutama pada jam sibuk atau kondisi lalu lintas tidak normal.

- **Layanan RoyalTrans**  memiliki median durasi perjalanan yang sedikit lebih panjang yaitu sekitar 75 menit, dengan IQR sebesar 45 menit, menunjukkan adanya rentang waktu tempuh antara 53 hingga 98 menit untuk sebagian besar perjalanan. Durasi yang lebih panjang ini wajar terjadi karena mayoritas rute RoyalTrans merupakan perjalanan jarak menengah hingga jauh, yang menghubungkan kawasan penyangga di luar Jakarta seperti Bekasi, Depok, dan Tangerang dengan pusat kota. Menariknya, jumlah outlier pada layanan ini sangat sedikit (hanya 2 kasus), menandakan bahwa RoyalTrans memiliki performa durasi yang cukup konsisten, sehingga cocok digunakan untuk mobilitas harian jarak jauh yang membutuhkan kepastian waktu.

- **Layanan Mikrotrans** juga memiliki median durasi sekitar 72 menit dengan IQR 45 menit, menunjukkan bahwa sebagian besar perjalanan berlangsung antara 50 hingga 95 menit. Namun demikian, ditemukan 9 kasus outlier dengan durasi yang jauh lebih lama dari normal. Hal ini bisa terjadi karena Mikrotrans melayani rute-rute pengumpan dalam kota dengan kondisi jalan yang beragam dan kerap padat, serta frekuensi pergerakan yang tidak seintens layanan utama. Selain itu, tidak adanya jalur khusus membuat layanan ini lebih mudah terdampak oleh kondisi lalu lintas lokal, terutama di area pemukiman padat atau pasar tradisional.

Untuk mendorong peningkatan jumlah perjalanan di akhir pekan, Transjakarta dapat menerapkan program loyalitas berupa tarif gratis setiap akhir pekan bagi penumpang aktif, yaitu mereka yang telah melakukan perjalanan lebih dari 15 hari dalam sebulan terakhir—pada 10 koridor dengan jumlah penumpang tertinggi saat weekend. Strategi ini tidak hanya mendorong penumpang aktif untuk tetap menggunakan layanan di akhir pekan, tetapi juga memotivasi penumpang lainnya agar lebih sering bepergian menggunakan Transjakarta demi memperoleh manfaat serupa. Dengan demikian, program ini diharapkan dapat memperluas keterlibatan pengguna setia sekaligus secara bertahap meningkatkan frekuensi perjalanan dan jumlah penumpang di akhir pekan.

### **Rekomendasi: Analisa Durasi Perjalanan**

Penguatan pengawasan ini akan membantu meminimalkan risiko keterlambatan dan menekan jumlah kasus durasi ekstrem.

Upaya ini perlu diperluas dan diintegrasikan dengan sistem prediksi keterlambatan berbasis data historis, agar penyesuaian operasional bisa dilakukan secara cepat terhadap rute-rute yang menunjukkan potensi keterlambatan tinggi.

Untuk meningkatkan kualitas layanan dan membantu penumpang dalam merencanakan perjalanan dengan lebih baik, perlu dilakukan pengembangan aplikasi Transjakarta dengan menambahkan fitur yang dapat menampilkan rentang durasi perjalanan berdasarkan data historis. Fitur ini akan memberikan estimasi waktu tempuh yang lebih akurat, termasuk variasi durasi akibat kemacetan atau kondisi cuaca, serta bermanfaat untuk meningkatkan transparansi layanan, mengurangi ketidakpastian penumpang, dan mendorong kepercayaan serta kepuasan pelanggan.
